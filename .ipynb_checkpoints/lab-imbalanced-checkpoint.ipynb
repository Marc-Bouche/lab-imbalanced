**Load the data**

In this challenge, we will be working with Credit Card Fraud dataset.

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv

Metadata

- **distance_from_home:** the distance from home where the transaction happened.
- **distance_from_last_transaction:** the distance from last transaction happened.
- **ratio_to_median_purchase_price:** Ratio of purchased price transaction to median purchase price.
- **repeat_retailer:** Is the transaction happened from same retailer.
- **used_chip:** Is the transaction through chip (credit card).
- **used_pin_number:** Is the transaction happened by using PIN number.
- **online_order:** Is the transaction an online order.
- **fraud:** Is the transaction fraudulent. **0=legit** -  **1=fraud**


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import math
%matplotlib inline

# Machine Learning libraries
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.utils import resample
from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix, f1_score


# Metrics
from sklearn.metrics import (
    accuracy_score, 
    confusion_matrix, 
    precision_score, 
    recall_score, 
    f1_score, 
    roc_auc_score, 
    mean_absolute_error, 
    mean_squared_error, 
    r2_score
)
```

```python
fraud = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv")
fraud.head()
```

### 1. What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?

```python
fraud['fraud'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the values
fraud_counts = fraud['fraud'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values, palette="viridis")
plt.title("Distribution of Fraud vs. Non-Fraud Cases")
plt.xlabel("Fraud (1) vs. Non-Fraud (0)")
plt.ylabel("Count")
plt.xticks([0, 1], ["Non-Fraud", "Fraud"])
plt.show()
```

### 2. Train a LogisticRegression.

```python
# Select relevant features and target
features = fraud.drop(columns=['fraud'])
target = fraud['fraud']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)

# Initialize the scaler and scale the training and test sets
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform on training data
X_test_scaled = scaler.transform(X_test)        # Only transform test data

# Initialize and train the logistic regression model
log_reg = LogisticRegression(max_iter=100)
log_reg.fit(X_train_scaled, y_train)

# Make predictions
pred1 = log_reg.predict(X_test_scaled)
```

### 3. Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.

```python
# Evaluate the model
accuracy = accuracy_score(y_test, pred1)
precision= precision_score(y_test, pred1)
recall = precision_score(y_test, pred1)
f1 = f1_score(y_test, pred1)
conf_matrix_lr = confusion_matrix(y_test, pred1)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("f1:", f1)

print("Confusion Matrix:\n", conf_matrix_lr)
```

```python
# Confusion Matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_lr, annot=True, fmt="d", cmap="Blues", cbar=False, 
            xticklabels=['Non-Fraud', 'Fraud'], 
            yticklabels=['Non-Fraud', 'Fraud'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Model Performance")
plt.show()
```

### 4. Run Oversample in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model?

```python
# Convert Training Data to DataFrame
train = pd.DataFrame(X_train_scaled, columns = X_train.columns)
train["fraud"] = y_train.values

#Separate Fraud and Non-Fraud Records
fraud = train[train["fraud"] == 1]
no_fraud = train[train["fraud"] == 0]

#Oversample Fraud Cases
yes_fraud_oversampled = resample(fraud, 
                                    replace=True, 
                                    n_samples = len(no_fraud),
                                    random_state=0)

#Combine Oversampled Fraud with Non-Fraud
train_over = pd.concat([yes_fraud_oversampled, no_fraud])
train_over
```

```python
# Plot visulization
survived_plt = train_over["fraud"].value_counts()
survived_plt.plot(kind="bar")
plt.show()
```

```python
X_train_over = train_over.drop(columns = ["fraud"])
y_train_over = train_over["fraud"]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_over, y_train_over)
```

```python
pred = log_reg.predict(X_test_scaled)
print(classification_report(y_pred = pred, y_true = y_test))
```

```python
# Visualization

# Generate classification report as a dictionary
report = classification_report(y_true=y_test, y_pred=pred, output_dict=True)
report_df = pd.DataFrame(report).T  # Convert to DataFrame and transpose for readability

# Filter only precision, recall, and f1-score columns
report_df = report_df[["precision", "recall", "f1-score"]]

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(report_df.iloc[:2], annot=True, cmap="Blues", fmt=".2f", cbar=True)
plt.title("Classification Report - Precision, Recall, F1-Score")
plt.xlabel("Metrics")
plt.ylabel("Class")
plt.show()
```

### 5. Now, run Undersample in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?

```python
no_fraud_undersampled = resample(no_fraud, 
                                    replace=False, 
                                    n_samples = len(fraud),
                                    random_state=0)
train_under = pd.concat([no_fraud_undersampled, fraud])

X_train_under1 = train_under.drop(columns = ["fraud"])
y_train_under1 = train_under["fraud"]
log_reg = LogisticRegression()
log_reg.fit(X_train_under1, y_train_under1)

#Evaluate the model 

pred1 = log_reg.predict(X_test_scaled)
print(classification_report(y_pred = pred1, y_true = y_test))
```

```python
# Visualization

# Generate classification report as a dictionary
report = classification_report(y_true=y_test, y_pred=pred1, output_dict=True)
report_df = pd.DataFrame(report).T  # Convert to DataFrame and transpose for readability

# Filter only precision, recall, and f1-score columns
report_df = report_df[["precision", "recall", "f1-score"]]

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(report_df.iloc[:2], annot=True, cmap="Blues", fmt=".2f", cbar=True)
plt.title("Classification Report - Precision, Recall, F1-Score")
plt.xlabel("Metrics")
plt.ylabel("Class")
plt.show()
```
